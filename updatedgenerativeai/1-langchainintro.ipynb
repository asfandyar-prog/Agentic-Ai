### Streaming & Batch:

# Streaming = the model sends the answer piece by piece as it thinks

nstead of waiting for the full response, you get tokens in real time.

Like:

WhatsApp “typing…”

ChatGPT text appearing gradually

Batching (in GenAI)
🧠 Idea (human-level intuition)

Batching = send many prompts at once instead of one-by-one

Like sending a list of questions in one request.

```python
responses=model.batch([
"what is quantum computing and quatum ai and relation between agi?",
"how airplanes fly?"])

for response in responses:
    print(response.content)
```
