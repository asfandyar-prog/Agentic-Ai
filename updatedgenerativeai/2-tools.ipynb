## Tools


Tools let an LLM call external functions or APIs instead of only generating text.

They are used when the model needs real data, calculations, or actions.

Examples: search the web, call a database, run Python code, use an API.

The LLM decides when to use a tool and sends structured input to it.

The tool runs, returns results, and the LLM uses them to answer correctly.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

model=init_chat_model("groq:llama-3.1-8b-instant")
response=model.invoke("give me one secret of genai")
print(response.content)
```

```python
from langchain.tools import tool
@tool
def  get_weather(location:str)->str:
    """get the weather in a given location"""
    return f"The weather in {location} is sunny"

model_with_tools = model.bind_tools([get_weather])
```

```python
res=model_with_tools.invoke("What is the weather of France?")
```

```python
print(res)
```

```python
res.tool_calls
```

# Tool Execution Loop

```python
messages=[{"role":"user","content":"what is the weather in France?"}]
ai_msg=model_with_tools.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result=get_weather.invoke(tool_call)
    messages.append(tool_result)
```

```python
final_res=model_with_tools.invoke(messages)
```

```python
print(final_res.text)
```
